**Analogías**

Las Temporal Neural Networks (TNN) trabajan con información que depende del tiempo. Para comprender mejor su funcionamiento, es útil compararlas con situaciones cotidianas que ayudan a visualizar cómo procesan y almacenan información.

**Una TNN como una persona con memoria**

Una TNN puede compararse con una persona que toma decisiones basándose no solo en lo que ocurre en este momento, sino también en lo que ha vivido antes.

Por ejemplo, al mantener una conversación, una persona entiende el significado de una frase porque recuerda lo que se dijo anteriormente. De forma similar, una TNN utiliza información pasada para interpretar correctamente la entrada actual y generar una salida coherente.

**Ventana temporal: “mirar los últimos minutos”**

Las TNN basadas en ventanas temporales funcionan como alguien que, antes de decidir algo, revisa lo ocurrido en los últimos minutos.

La red recibe varios valores consecutivos del pasado junto con el valor actual, formando un pequeño historial reciente. Esto le permite detectar patrones que no serían visibles si solo se analizara un único instante de tiempo.

En este enfoque, la red no tiene memoria a largo plazo, sino que observa únicamente un tramo fijo del pasado.

**Estados internos: “recordar el contexto”**

En las TNN con estados internos, la red posee una memoria dinámica que se actualiza constantemente. Esta memoria representa el contexto acumulado de la información previa.

Es similar a cómo una persona recuerda el tema general de una conversación, incluso si no recuerda cada frase exacta. La decisión actual se toma teniendo en cuenta ese contexto interno, no solo los datos más recientes.

Este tipo de TNN es especialmente útil cuando la influencia del pasado no está limitada a una ventana temporal fija.
