**-- ¿Qué son las TNN? --**

Las **Temporal Neuronal Networks (TNN)** son un conjunto de redes neuronales artificiales que procesan información cuya estructura depende del **tiempo**, es decir, de datos donde el significado de la observación depende no solo del valor presente, sino tambien del historial de estados previos en la secuencia temporal. A diferencia de otras redes neuronales tradicionales, que toman como independientes el valor de sus entradas, las TNN tienen mecanismos que permiten incorporar el concepto del tiempo en su estructura interna, por lo que las TNN agrupan enfoques que permiten capturar relaciones temporales.

A diferencia de las redes neuronales convencionales, que procesan entradas estáticas sin considerar el orden temporal, las TNN incorporan mecanismos que permiten capturar dependencias temporales, ya sea mediante retardos en las entradas o a través de estados internos con memoria. De esta manera, las TNN permiten modelar relaciones donde el pasado influye de forma directa en el presente y en el futuro.

Desde el punto de vista del aprendizaje automático, las TNN no corresponden a una única arquitectura específica, sino a una familia de enfoques que comparten un objetivo común: representar y aprender patrones que se desarrollan a lo largo del tiempo. Dentro de este conjunto se incluyen modelos como las redes con retardos temporales y las redes neuronales recurrentes, ampliamente utilizadas en problemas secuenciales.

**EJEMPLOS DE USO**

Las TNN se aplican principalmente en problemas donde los datos presentan una estructura temporal clara y el comportamiento futuro depende del historial del sistema. Algunos ejemplos representativos son:

**- Predicción de valores futuros en series temporales**, donde se busca estimar comportamientos a corto o mediano plazo a partir de datos históricos.
*(Ejemplo: predicción de demanda eléctrica utilizando registros históricos de consumo para la estimación de carga futura).*

**- Modelado y control de sistemas dinámicos**, en los que el estado actual del sistema depende de su evolución previa, siendo común su uso en automatización y robótica.
*(Ejemplo: control de un brazo robótico, donde la posición actual depende de la velocidad y posición en instantes anteriores).*
